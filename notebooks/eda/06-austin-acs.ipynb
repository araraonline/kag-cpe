## Poverty

*Margin of Error is usually big*

```python
path = '../../data/departments/37-00027/external/ACS/ACS_16_5YR_S1701_with_ann.csv'
poverty = pd.read_csv(path, skiprows=[1])
poverty_names = pd.read_csv(path, nrows=1).T
```

```python
poverty.head()
```

```python
poverty.isnull().values.sum()
# no nulls
```
